```python
df2 = pd.read_csv("Rat_Sightings.csv", low_memory=False)
df2.head()
```

```python
df1 = df1.rename(columns={'BOROUGH': 'borough'})
df1['borough'] = df1['borough'].str.lower()
df1
```

```python
df2 = df2.rename(columns={'Borough': 'borough'})
df2['borough'] = df2['borough'].str.lower()
df2
```

```python
df2.dtypes
```

```python
df1['borough'] = df1['borough'].fillna(0)
df2['borough'] = df2['borough'].fillna(0)
```

```python
df1['borough'].value_counts()
```

```python
df2['borough'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
plt.plot(df1['borough'].value_counts(), marker="o", c='tab:red', ls='--', lw=1, ms=8, mew=3, mec='red')#, alpha=.7)
plt.xlabel('Boroughs')
plt.ylabel('Number of Inspections')
plt.title("Amount of inspection, per borough, since 2010")
plt.legend(['Inspections']);
```

```python
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
plt.plot(df2['borough'].value_counts(), marker="o", c='tab:blue', ls='--', lw=1, ms=8, mew=3, mec='blue')
plt.xlabel('Boroughs')
plt.ylabel('Number of Calls')
plt.title("Amount of calls, per borough, since 2010")
plt.legend(['Calls to #311']);
```

```python
df1_inspect = df1['borough'].value_counts().reset_index()
df2_calls = df2['borough'].value_counts().reset_index()
```

```python
df = df1_inspect.merge(df2_calls, on=['borough'], how='outer') 
df
```

```python
df_altair = df.melt(id_vars='borough')
df_altair
```

```python
import altair as alt

alt.Chart(df_altair).mark_bar().encode(
    x='borough',
    y='value',
    color='variable',
    
)
```
