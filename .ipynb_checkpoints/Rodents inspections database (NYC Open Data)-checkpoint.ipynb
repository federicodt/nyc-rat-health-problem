```python
import requests
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from pywaffle import Waffle
import matplotlib.pylab as pylab
from matplotlib.ticker import FuncFormatter
%matplotlib inline
import numpy as np
import altair as alt
from vega_datasets import data
import seaborn as sns
```

```python
df1 = pd.read_csv("NYC Open Databases\Rodent_Inspection.csv", low_memory=False)
df1.head()
```

```python
colors = ['#F03E3E', '#4DABF7', '#4DABF7', '#4DABF7', '#4DABF7', '#4DABF7', '#4DABF7']

ax = df1['RESULT'].value_counts().head(10).plot(kind='bar', color=colors, figsize=(15,12))

ax.set_title('Inspections results', pad=40)
ax.title.set_fontsize(15)

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.title.set_fontsize(33)
ax.title.set_fontname('Arial')

ax.tick_params(which='both', left=False, bottom=False, grid_alpha=0.9)
ax.grid(True, axis='y', color='#212529', linestyle=':', linewidth=0.9)

ax.set_axisbelow(True)

plt.ylabel("Number of inspections")
plt.xlabel("Inspections results")
plt.xticks(rotation=45, horizontalalignment="center")

plt.rcParams.update({    
    'ps.fonttype': 42,
    'figure.facecolor': '#F8F9FA', # colors the outside
    'axes.facecolor': '#F8F9FA', # colors the inside 
    'font.size': 15,
    'text.color': '#212529' 
})
```

```python
bars_inspections_results = alt.Chart(df1['RESULT'].value_counts().head(10).reset_index()).mark_bar(size=75).encode(
    x = alt.X("RESULT", title="Inspections results", sort=None, axis=alt.Axis(labelAngle=-45, labelFontWeight='bold')),
    y = alt.Y("count", title="Number of inspections"),    
    color= alt.condition(
        alt.datum.count > 1000000, 
        alt.value("red"),
        alt.value("steelblue"))
)

text_inspections_results = bars_inspections_results.mark_text(
    align='center',
    baseline='middle',
    dx=0,  # Nudges text to right so it doesn't appear on top of the bar
    dy=-10,
    fontStyle='bold',
    fontSize= 15
).encode(text='count')

inspections_results = (bars_inspections_results + text_inspections_results).properties(width = 'container', height = 600, 
    title = "Inspections made for rodents (2010 - June 25, 2023)").configure_title(
    fontSize=20).configure_view(strokeWidth=0).configure_axis(labelFontSize=13,
    titleFontSize=18, grid=False)

inspections_results
```

```python
BR = df1[['BOROUGH', 'RESULT']]
BR
```

```python
boro_inspect = BR.groupby(['BOROUGH', BR['RESULT']=='Passed']).size()#.groupby(level=1).max()
boro_inspect
```

```python
b_inspect = boro_inspect.to_frame(name='count')
b_inspect
```

```python
b_inspect.reset_index(inplace=True)
b_inspect
```

```python
b_inspects = b_inspect.replace({
    True: 'Passed',
    False: 'Failed'
})
```

```python
import altair as alt
from vega_datasets import data

bars_boroughs_inspections = alt.Chart(b_inspects).mark_bar(size=75).encode(
    x = alt.X('BOROUGH', title="Boroughs", sort=None, axis=alt.Axis(labelAngle=-45, labelFontWeight='bold')),
    y = alt.Y('count', title="Number of inspections"),
    color = 'RESULT',
    order=alt.Order('RESULT:N', sort='descending')
)

text_boroughs_inspections = bars_boroughs_inspections.mark_text(
align='center',
    baseline='middle',
    dx=0,  # Nudges text to right so it doesn't appear on top of the bar
    dy=-10,
    fontStyle='bold',
    fontSize= 15
)

inspections_by_boroughs = (bars_boroughs_inspections + text_boroughs_inspections).properties(
    width = 'container', height = 600, 
    title = "Inspections made for rodents by Boroughs (2010 - June 25, 2023)").configure_title(
    fontSize=20).configure_view(strokeWidth=0).configure_axis(labelFontSize=13, titleFontSize=15, grid=False).configure_legend(titleColor='black', 
    titleFontSize=13, labelFontSize=13, titleFontWeight='bold', symbolSize=200)

config = alt.layer(bars_boroughs_inspections)

inspections_by_boroughs
```
